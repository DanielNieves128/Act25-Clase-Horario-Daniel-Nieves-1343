```python
class Horario:
    def __init__(self, dia="", hora_inicio="", hora_fin=""):
        self.dia = dia
        self.hora_inicio = hora_inicio
        self.hora_fin = hora_fin

    def leerdatos(self, numero_objeto):
        print(f"\n--- Ingresando datos para el Horario {numero_objeto} ---")
        self.dia = input("Ingrese el día: ")
        self.hora_inicio = input("Ingrese la hora de inicio (HH:MM): ")
        self.hora_fin = input("Ingrese la hora de fin (HH:MM): ")

    def mostrardatos(self):
        print(f"Día: {self.dia:10} | Turno: {self.hora_inicio} a {self.hora_fin}")

# --- Inicio del Programa Interactivo ---

print("="*40)
print("SISTEMA DE GESTIÓN DE HORARIOS")
print("="*40)

# 1. Creación de los 2 objetos (instancias vacías)
horario1 = Horario()
horario2 = Horario()

# 2. Uso de la función leerdatos() para cada objeto
horario1.leerdatos(1)
horario2.leerdatos(2)

# 3. Uso de la función mostrardatos() para visualizar los resultados
print("\n" + "="*40)
print("RESUMEN DE HORARIOS REGISTRADOS")
print("="*40)
horario1.mostrardatos()
horario2.mostrardatos()

print("\n" + "="*40)
print("Programa mejorado sobre la base de: Daniel Nieves NL=41 NC=1343")
print("="*40)
```
